```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset= pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv', skiprows=1)
```

```python
dataset.info()
```

Data Cleaning

```python
#missing values
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converted into two sets based on Region from 122th index, we can make a new column based on the Region

1: "Bejaia Region Dataset"
2: "Sidi-Bel Abbes Region Dataset"

Add New coloumn with Region

```python
df = dataset.copy()
```

```python
df.loc[0:122, 'Region'] = 0
df.loc[122:,'Region'] = 1
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.isnull().sum()
```

```python
# Removing null values by imputing with mean
df.dropna(inplace=True)
```

```python
df=df.reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df= df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
# fix spaces in column names
df.columns = df.columns.str.strip()
df.columns
```

```python
# change the required coloumns as int type
df[['month','day','year','Temperature', 'RH', 'Ws']]= df[['month','day','year','Temperature', 'RH', 'Ws']].astype(int) 
df.info() 
```

```python
# changing other columns to float type
objects= [features for features in df.columns if df[features].dtypes == 'object']
```

```python
for i in objects:
    if i != 'Classes':
        df[i]= df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
## lets save the cleaned data
df.to_csv('Algerian_forest_fires_cleaned.csv', index=False)
```
